```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#import data set
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
house_price_dataframe.head()
```

```python
#adding the target column(price) which is the label to the datafram
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
#check for null values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures
house_price_dataframe.describe()
```

```python
#check for negative nd positive correlation
correlation = house_price_dataframe.corr()
```

```python
print(correlation)
house_price_dataframe.dropna(inplace = True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar = True, square=True, fmt='.01f', annot=True, annot_kws={'size':8}, cmap = 'Blues')
```

```python
x = house_price_dataframe.drop(columns=['price'], axis = 1)
y = house_price_dataframe['price']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, train_size = .8, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#train model: XGBoost regressor
#load the model
model = XGBRegressor(missing = 1)
```

```python
model.fit(x_train,y_train)
```

```python
training_prediction = model.predict(x_train) #contains the prices predicted for the training data
```

```python
print(training_prediction)
```

```python
# R squared error for regression problems and accuracy score for classification problems
score_1 = metrics.r2_score(y_train, training_prediction)
print("R squared error: ", score_1)
#mean absolute is another accuracy check method for "Regression problems"
score_2 = metrics.mean_absolute_error(y_train, training_prediction)
print("Mean absolute error: ", score_2)
```

```python
#Checking accuracy for testing data
test_prediction = model.predict(x_test)
score1 = metrics.r2_score(y_test, test_prediction)
score2 = metrics.mean_absolute_error(y_test, test_prediction)
print("R squared error: ", score1)
print("Mean absolute error: ", score2)
```

```python
plt.scatter(y_test, test_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
